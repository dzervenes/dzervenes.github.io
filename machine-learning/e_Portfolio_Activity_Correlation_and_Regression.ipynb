**Covariance and Pearson Correlation**

The notebook titled *Unit03 Ex1 covariance_pearson_correlation*.ipynb was executed in a separate window using the appropriate Jupyter environment. The code provided as part of this exercise has been included below for reference and documentation.

```python

# calculate the Pearson's correlation between two variables
from numpy import mean
from numpy import std
from numpy import cov
from numpy.random import randn
from numpy.random import seed
from matplotlib import pyplot as plt
import seaborn as sns

from scipy.stats import pearsonr
# seed random number generator
seed(1)

# prepare data
data1 = 20 * randn(1000) + 100
data2 = data1 + (10 * randn(1000) + 50)

# calculate covariance matrix
covariance = cov(data1, data2)

# calculate Pearson's correlation
corr, _ = pearsonr(data1, data2)

# plot
plt.scatter(data1, data2)
plt.show()

# summarize
print('data1: mean=%.3f stdv=%.3f' % (mean(data1), std(data1)))
print('data2: mean=%.3f stdv=%.3f' % (mean(data2), std(data2)))
print('Covariance: %.3f' % covariance[0][1])
print('Pearsons correlation: %.3f' % corr)
```

**Analysis**

This dataset demonstrates a strong positive linear relationship, highlighted by a Pearson's correlation of  = 0.888. Data1 (mean = 100.776, stdv=19.620) and data2 (mean = 151.050, stdv =
22.358) exhibit moderate variability.

The covariance of 389.755 further supports the consistent upward trend observed in the scatter plot.

The notebook titled *Unit03_Ex3 multiple_linear_regressions.ipynb* was executed in a separate window using the appropriate Jupyter environment. The code provided as part of this exercise has been included below for reference and documentation.

```python
import pandas
from sklearn import linear_model

df = pandas.read_csv("cars.csv")

X = df[['Weight', 'Volume']]
y = df['CO2']

regr = linear_model.LinearRegression()
regr.fit(X, y)

#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1300cm3:
predictedCO2 = regr.predict([[2300, 1300]])

print(predictedCO2)

print(regr.coef_)

predictedCO2 = regr.predict([[3300, 1300]])

print(predictedCO2)
```

This code initially calculates the predicted CO2 emissions for a baseline scenario, resulting in a value of 107.21. Additionally, it evaluates the impact of increasing the vehicle's weight, leading to an updated CO2 prediction of 114.76. In the subsequent experiment, I will analyse the effect of increasing the volume alone and then observe the combined impact of increasing both the weight and volume simultaneously.

```python
#predict the CO2 emission of a car where the weight is 2300kg, and the volume is 1800cm3:
predictedCO2 = regr.predict([[2300, 1800]])

print(predictedCO2)

#predict the CO2 emission of a car where the weight is 3300kg, and the volume is 1800cm3:

predictedCO2 = regr.predict([[3300, 1800]])

print(predictedCO2)
```

**Analysis**

The analysis begins with a baseline CO2 emission prediction of 107.21 for a vehicle weighing 2300kg with a volume of 1300cm. Increasing the weight to 3300kg raises the prediction to 114.76, while increasing the volume to 1800cm at the same weight results in 111.11. A simultaneous increase in weight and volume yields the highest predicted emission of 118.66, illustrating the combined impact of these variables on CO2 emissions.

This dataset demonstrates a strong polynomial relationship, as indicated by an R2 value of 0.943, reflecting a high degree of fit between the model and the data. The curve captures the non-linear trend accurately, and the predicted speed for x=17, aligning well with the observed pattern.

1. Reducing polynomial degree

The polynomial degree was reduced from 3 to 2 to simplify the model and assess whether a lower-degree polynomial could effectively capture the data's trends. This change balances model simplicity and fit quality, aiming to reduce overfitting while maintaining accuracy. The impact was evaluated by comparing the regression curve visually and through the R2 score.

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100]

#NumPy has a method that lets us make a polynomial model
mymodel = numpy.poly1d(numpy.polyfit(x, y, 2))

#specify how the line will display, we start at position 1, and end at position 22
myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()

from sklearn.metrics import r2_score

mymodel = numpy.poly1d(numpy.polyfit(x, y, 2))

print(r2_score(y, mymodel(x)))

#Predict future values

mymodel = numpy.poly1d(numpy.polyfit(x, y, 2))

speed = mymodel(17)
print(speed)
```

**Analysis**

Reducing the polynomial degree from 3 to 2 resulted in a weaker fit, with R2
decreasing from 0.943 to 0.760. This indicates that the quadratic model is less able to capture the nuances of the dataset's curvature compared to the cubic model.

The predicted speed also shifted, dropping from 88.87 in the cubic model to
81.18 in the quadratic model, reflecting the less precise alignment of the curve to the data points. Visually, the second-degree polynomial fails to capture the full variability in the data, particularly in areas with sharper changes, whereas the third-degree polynomial provides a much closer fit to the observed trend.

2. Adjusting existing data points

All y-values below 70 were adjusted to 70 to standardize the dataset and reduce variability. This modification aims to assess its impact on the regression model's curve, R2 score, and predictions.

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22]
y = [100,90,80,70,70,70,70,70,70,70,75,76,78,79,90,99,99,100]

#NumPy has a method that lets us make a polynomial model
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

#specify how the line will display, we start at position 1, and end at position 22
myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()

from sklearn.metrics import r2_score

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))

#Predict future values

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

speed = mymodel(17)
print(speed)
```

Adjusting all y-values below 70 to a standardized value of 70 resulted in a stronger fit, with R2 increasing from 0.943 to 0.958. This modification reduced variability in the dataset, making the model more consistent and aligned with the adjusted data.

The predicted speed for
x=17 shifted slightly from 88.87 in the original dataset to
86.65, reflecting the influence of the adjustedy-values on the regression curve. Visually, the new curve appears smoother and better conforms to the adjusted dataset, highlighting the effect of reducing variability on improving the regression model's fit and consistency.

```python
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22,75]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100,175]

#NumPy has a method that lets us make a polynomial model
mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

#specify how the line will display, we start at position 1, and end at position 22
myline = numpy.linspace(1, 22, 100)

plt.scatter(x, y)
plt.plot(myline, mymodel(myline))
plt.show()

from sklearn.metrics import r2_score

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

print(r2_score(y, mymodel(x)))

#Predict future values

mymodel = numpy.poly1d(numpy.polyfit(x, y, 3))

speed = mymodel(17)
print(speed)
```

**Analysis**

Adding an outlier to the dataset slightly reduced the R2 value from 0.943 to 0.942, indicating a minimal decrease in the model's fit. The predicted speed for
x=17 also shifted downward from 88.87 to 82.34, reflecting the influence of the outlier on the regression model.

Visually, the scatter plot shows the outlier significantly skewing the curve, particularly at the higher end of x-values, which distorts the natural trend observed in the original dataset. This highlights the sensitivity of polynomial regression models to extreme values, as even a single outlier can affect the fit, predictions, and overall trend.
